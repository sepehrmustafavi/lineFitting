```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate 10 random data points
np.random.seed(42)  # For reproducibility
x = np.random.rand(10)
y = 2 * x + 1 + 0.1 * np.random.randn(10)  # A line with some noise

# Initial guess for slope and intercept
m = 0
b = 0

# Learning rate (adjust as needed)
learning_rate = 0.01

# Number of iterations (adjust as needed)
num_iterations = 1000

# Perform LMS optimization
for _ in range(num_iterations):
    y_pred = m * x + b
    error = y - y_pred

    # Update parameters using LMS update rule
    m += learning_rate * np.mean(error * x)
    b += learning_rate * np.mean(error)

# Calculate fit error
fit_error = np.sum(error**2)

# Plot the data points and the fitted line
plt.scatter(x, y, label='Actual data points')
plt.plot(x, y_pred, color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"Fit error: {fit_error}")
```

```python
# Generate 10 random data points
np.random.seed(42)  # For reproducibility
x = np.random.rand(10)
y = 2 * x**2 + 0.5 * x + 1 + 0.1 * np.random.randn(10)  # A quadratic curve with some noise

# Stack the data points into a 2D array
data_points = np.column_stack((x, y))

# Initialize parameters for a higher-degree polynomial (ax^2 + bx + c)
a, b, c = 0, 0, 0

# Learning rate (adjust as needed)
learning_rate = 0.1

# Number of iterations (adjust as needed)
num_iterations = 10000

# Perform LMS optimization for the higher-degree polynomial
for _ in range(num_iterations):
    y_pred = a * x**2 + b * x + c
    error = y - y_pred

    # Update parameters using LMS update rule
    a += learning_rate * np.mean(error * x**2)
    b += learning_rate * np.mean(error * x)
    c += learning_rate * np.mean(error)

# Calculate fit error for the higher-degree polynomial
fit_error = np.sum(error**2)

# Plot the data points and the fitted higher-degree polynomial
plt.scatter(x, y, label='Actual data points')
x_curve = np.linspace(min(x), max(x), 100)
y_curve = a * x_curve**2 + b * x_curve + c
plt.plot(x_curve, y_curve, color='red', label='Fitted higher-degree polynomial')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Print parameters and fit error for the higher-degree polynomial
print(f"Higher-Degree Polynomial - Coefficients: a={a}, b={b}, c={c}")
print(f"Fit error: {fit_error}")
```
